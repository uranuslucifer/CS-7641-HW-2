```python
import numpy as np
import pandas as pd
import mlrose_hiive
import time
import matplotlib.pyplot as plt
```

### Four Peaks Problem (4PP)

```python
fitness = mlrose_hiive.FourPeaks(t_pct=0.15)
problem = mlrose_hiive.DiscreteOpt(length=100, fitness_fn=fitness, maximize=True, max_val=2)
```

```python
# RHC
start = time.time()
rhc_best_state, rhc_best_fitness, rhc_fitness_curve = mlrose_hiive.random_hill_climb(problem, restarts = 50, max_attempts = 300, random_state = 36, max_iters = 300, curve = True)
end = time.time()
RHC_Time_4PP = end-start
print("RHC Time: ", RHC_Time_4PP)
```

```python
# SA
start = time.time()
sa_best_state, sa_best_fitness, sa_fitness_curve = mlrose_hiive.simulated_annealing(problem, schedule = mlrose_hiive.ExpDecay(init_temp=1, exp_const=0.001, min_temp=0.001), 
                                                                                    max_attempts = 300, random_state = 36, max_iters = 300, curve = True)
end = time.time()
SA_Time_4PP = end-start
print("SA Time: ", SA_Time_4PP)
```

```python
# GA
start = time.time()
ga_best_state, ga_best_fitness, ga_fitness_curve = mlrose_hiive.genetic_alg(problem, pop_size=100, mutation_prob = 0.4, max_attempts = 300, random_state = 36, max_iters = 300, curve = True)
end = time.time()
GA_Time_4PP = end-start
print("GA Time: ", GA_Time_4PP)
```

```python
# MIMIC
start = time.time()
mimic_best_state, mimic_best_fitness, mimic_fitness_curve = mlrose_hiive.mimic(problem, pop_size = 200, keep_pct = 0.2, max_attempts = 300, random_state = 36, max_iters = 300, curve = True)
end = time.time()
MIMIC_Time_4PP = end-start
print("MIMIC Time: ", MIMIC_Time_4PP)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 301), rhc_fitness_curve[:,0], label = 'RHC')
plt.plot(range(1, 301), sa_fitness_curve[:,0], label = 'SA')
plt.plot(range(1, 301), ga_fitness_curve[:,0], label = 'GA')
plt.plot(range(1, 301), mimic_fitness_curve[:,0], label = 'MIMIC')
plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.title('Best Fitness vs Iterations for Four Peaks Problem')
plt.xticks(np.arange(0, 301, 50))
plt.legend(loc = 'upper right')
plt.show()
```

### N-Queens Problem (NQP)

```python
from mlrose_hiive import QueensGenerator
problem = QueensGenerator().generate(seed=123456, size=50)
```

```python
# RHC
start = time.time()
rhc_best_state, rhc_best_fitness, rhc_fitness_curve = mlrose_hiive.random_hill_climb(problem, 
                                                                                    max_attempts = 300, 
                                                                                    max_iters=300, 
                                                                                    curve=True, 
                                                                                    random_state=36,
                                                                                    restarts=50)
end = time.time()
RHC_Time_NQP = end-start
print("RHC Time: ", RHC_Time_NQP)
```

```python
# SA
start = time.time()
sa_best_state, sa_best_fitness, sa_fitness_curve = mlrose_hiive.simulated_annealing(
                                                    problem, 
                                                    max_attempts=300, 
                                                    max_iters=300, 
                                                    curve=True, 
                                                    random_state=36,
                                                    schedule=mlrose_hiive.ExpDecay(init_temp=30, exp_const=0.1, min_temp=0.5))
end = time.time()
SA_Time_NQP = end-start
print("SA Time: ", SA_Time_NQP)
```

```python
# GA
start = time.time()
a_best_state, ga_best_fitness, ga_fitness_curve = mlrose_hiive.genetic_alg(
                                                    problem, 
                                                    max_attempts=300, 
                                                    max_iters=300, 
                                                    curve=True, 
                                                    random_state=36,
                                                    pop_size=100,
                                                    mutation_prob=0.5)
end = time.time()
GA_Time_NQP = end-start
print("GA Time: ", GA_Time_NQP)
```

```python
# MIMIC
start = time.time()
mimic_best_state, mimic_best_fitness, mimic_fitness_curve = mlrose_hiive.mimic(
                                                            problem, 
                                                            max_attempts = 300, 
                                                            max_iters = 300,  
                                                            curve = True, 
                                                            random_state = 36,
                                                            pop_size=300,
                                                            keep_pct=0.3)
end = time.time()
MIMIC_Time_NQP = end-start
print("MIMIC Time: ", MIMIC_Time_NQP)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 301), rhc_fitness_curve[:,0], label = 'RHC')
plt.plot(range(1, 301), sa_fitness_curve[:,0], label = 'SA')
plt.plot(range(1, 301), ga_fitness_curve[:,0], label = 'GA')
plt.plot(range(1, 301), mimic_fitness_curve[:,0], label = 'MIMIC')
plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.title('Best Fitness vs Iterations for N-Queens Problem')
plt.xticks(np.arange(0, 301, 50))
plt.legend()
plt.show()
```
